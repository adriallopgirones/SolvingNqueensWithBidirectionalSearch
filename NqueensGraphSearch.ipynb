```python
import copy

def is_goal_puzzle(board):
    #Converting board in a flat list to make comparisions easier
    flat_board = [item for sublist in board for item in sublist]
    #We don't need the 0 to check if is a goal
    flat_board.remove(0)
    copy_flat_board = copy.copy(flat_board)
    
    #If the list is ordered the puzzle is solved
    if copy_flat_board == sorted(copy_flat_board) or copy_flat_board == sorted(copy_flat_board, reverse=True):
        return True
    else:
        return False

board = [
[8,0,6],
[5,4,7],
[2,3,1]
]

is_goal_puzzle(board)
```
